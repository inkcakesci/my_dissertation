# Baseline Periods Runner

This notebook runs `src.traditional.run_baseline_periods` in a cross-platform way.
It auto-detects the project root, then launches the module with the active Python interpreter.


```python
from pathlib import Path
import subprocess
import sys


def find_project_root(start: Path) -> Path:
    start = start.resolve()
    for candidate in [start, *start.parents]:
        if (candidate / "src").is_dir() and (candidate / "prepare_kepler_mcquillan.py").exists():
            return candidate
    raise FileNotFoundError(
        "Could not locate the project root. Expected a directory containing 'src/' and 'prepare_kepler_mcquillan.py'."
    )


PROJECT_ROOT = find_project_root(Path.cwd())
print(f"[INFO] Project root: {PROJECT_ROOT}")

MAX_TARGETS = 1000
SAVE_EVERY = 25
SKIP_GPS = False
SKIP_QPGP = False

cmd = [
    sys.executable,
    "-m",
    "src.traditional.run_baseline_periods",
    "--max-targets", str(MAX_TARGETS),
    "--save-every", str(SAVE_EVERY),
]
if SKIP_GPS:
    cmd.append("--skip-gps")
if SKIP_QPGP:
    cmd.append("--skip-qpgp")

print("[INFO] Running command:")
print(" ".join(cmd))

process = subprocess.Popen(
    cmd,
    cwd=PROJECT_ROOT,
    text=True,
    stdout=subprocess.PIPE,
    stderr=subprocess.STDOUT,
    bufsize=1,
)

assert process.stdout is not None
for line in process.stdout:
    print(line, end="")

return_code = process.wait()
if return_code != 0:
    raise RuntimeError(f"Baseline run failed with exit code {return_code}.")
```
