```python
import numpy as np

def validate_npz(path: Path) -> bool:
    """Check whether a .npz is loadable and contains required arrays."""
    if not path.exists():
        return False
    try:
        with np.load(path) as data:
            required_keys = ("time", "flux", "flux_err")
            for key in required_keys:
                if key not in data:
                    print(f"[WARN] {path} missing key '{key}'")
                    return False
                arr = data[key]
                if arr.size == 0:
                    print(f"[WARN] {path} key '{key}' has zero length")
                    return False
                if not np.isfinite(arr).any():
                    print(f"[WARN] {path} key '{key}' has no finite values")
                    return False
    except Exception as exc:
        print(f"[WARN] Failed to load npz ({path}): {exc}")
        return False

    return True
```

```python
lc_dir = PROJECT_ROOT / "data" / "kepler" / "lightcurves"
if not lc_dir.exists():
    print(f"[WARN] Directory not found: {lc_dir}")

npz_paths = sorted(lc_dir.glob("*.npz"))
valid_paths = []
invalid_paths = []

for path in npz_paths:
    if validate_npz(path):
        valid_paths.append(path)
    else:
        invalid_paths.append(path)

print(f"Found {len(npz_paths)} .npz files in {lc_dir}")
print(f"Valid .npz count: {len(valid_paths)}")

if invalid_paths:
    print("Invalid files:")
    for path in invalid_paths:
        print(f"- {path.name}")
```
