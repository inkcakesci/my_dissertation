# LightPred Training Visualization

This notebook reads the training log generated by `train_lightpred.py` and plots loss/MAE.

```python
from pathlib import Path
import pandas as pd

log_path = Path("data/kepler/models/lightpred_train_log.csv")
if not log_path.exists():
    raise FileNotFoundError(f"Log file not found: {log_path}")

df = pd.read_csv(log_path)
df.head()
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].plot(df['epoch'], df['train_loss'], label='train')
axes[0].plot(df['epoch'], df['val_loss'], label='val')
axes[0].set_title('Gaussian NLL')
axes[0].set_xlabel('epoch')
axes[0].set_ylabel('loss')
axes[0].legend()

axes[1].plot(df['epoch'], df['train_mae'], label='train')
axes[1].plot(df['epoch'], df['val_mae'], label='val')
axes[1].set_title('Period MAE (days)')
axes[1].set_xlabel('epoch')
axes[1].set_ylabel('mae')
axes[1].legend()

plt.tight_layout()
plt.show()
```
